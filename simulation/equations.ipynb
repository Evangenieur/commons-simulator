```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## The a_rv equation generates a distribution where 50% of the numbers are < 0.25

```python
a_rv = []
for i in range(4000):
    rv = np.random.rand()
    a_rv.append(1-4*(1-rv)*rv)
df = pd.DataFrame({"a_rv": a_rv})
plot = sns.distplot(df)
plt.show()
```

## expon.rvs() gives a distribution with a minimum number and a configurable standard deviation with loc and scale

```python
from scipy.stats import expon, gamma

e_rv = [expon.rvs(loc=0, scale=2) for i in range(100)]
df = pd.DataFrame({"e_rv": e_rv})
plot = sns.distplot(e_rv)
plt.show()
```
